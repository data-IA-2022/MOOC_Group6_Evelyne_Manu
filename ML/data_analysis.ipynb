# Niveau d'études des utilisateurs

Il n'y a pas suffisamment de données pour faire une analyse par notes obtenues. Tout ce qu'on peut dire, c'est que la majorité des utilisateurs ont fait des études supérieures ('m' pour master ou 'b' pour bachelor). Par contre, il est surprenant que la proportion de doctotants ("p" pour PhD) soit aussi faible. Ils considèrent peut-être que le niveau des cours n'est pas assez pointu pour eux.

```python
# Choix de la palette de couleurs.
couleurs =['#636EFA', '#EF553B', '#00CC96', '#FF6692', '#AB63FA', '#19D3F3', '#B6E880', '#FF97FF', '#FECB52']
# Pie chart du niveau d'études pour l'ensemble des utilisateurs.
educ = df['level_of_education'].dropna()
educ = educ.replace(['none', 'None'], ['other', 'other'])
fig = px.pie(educ, names='level_of_education', color_discrete_sequence=couleurs, width=1000, height=1000, hole=.3, title="Niveau d'études des utilisateurs")
fig.update_traces(texttemplate='%{percent:.1%}')
fig.show()
```

# Répartition des notes en mode Log

Le mode Log permet de visualiser l'augmentation des notes au-dessus de 0.6. Cela semble indiquer que les tests ne sont pas particulièrement difficiles, mais demandent surtout de faire l'effort d'aller jusqu'au bout de la formation.

```python
# Histogramme des notes en mode Log.
fig = px.histogram(df, x='grade',  nbins=10, log_y=True, title='Répartition des notes en mode Log')
fig.show()
```

# Réussite par genre

Il y a énormément de données manquantes (en bleu). Tout ce qu'on peut dire, c'est que la très faible proportion de femmes ('f') parmi les participants se retrouve parmi les personnes qui réussissent la formation: une fine ligne verte à peine visible...

```python
# Répartition par genre pour la réussite et l'échec.
dfg = df
dfg['gender'] = dfg['gender'].replace('o', '')
def transf(grade):
  if grade < 0.5:
    return 'fail'
  else:
    return 'pass'
dfg['grade'] = dfg['grade'].apply(transf)
fig = px.histogram(dfg, x='grade', color='gender', opacity=.8, title='Réussite par genre',width=1000, height=1000)
fig.show()
```

# Répartition de la polarité par genre

Globalement, il y a un nombre significatif de personnes très polarisées, mais la majorité ont exprimé un sentiment raisonnablement neutre vis-à-vis des cours ou de leur expérience de la formation.

Contrairement aux préjugés, les hommes ne semblent pas plus critiques que les femmes. Et les femmes ne semblent pas davantage bienveillantes...
 

```python
# Histogramme de la polarité en fonction du genre.
fig = px.histogram(df, x='body', color='gender', nbins=20, log_y=False, title='Répartition de la polarité par genre')
fig.show()
```
